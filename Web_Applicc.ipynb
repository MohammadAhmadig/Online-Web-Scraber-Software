```python
#import selenium compnents, urllib, beautiful soup
from bs4 import BeautifulSoup
from selenium import webdriver
#import urllib.request
import smtplib
import winsound
import time
from datetime import datetime
from PIL import Image  

 
browser = webdriver.Chrome(r'chromedriver')
browser2 = webdriver.Chrome(r'chromedriver')
#open the browser with the URL
#a browser windows will appear for a little while
val1 = input("Addresse konoonie site pars90 ra bedoone http://www. vared konid: ") 
row1 = input("radife morede nazar dar site pars90 ra vared konid: ") 
val2 = input("Addresse konoonie site 1xbet ra bedoone https://www. vared konid: ") 
row2 = input("radife morede nazar dar site 1xbet ra vared konid: ") 
#url = "http://www.irpars90.com/Sportbook/InPlay"
#url = "https://1xirsp96.com/"
url = "http://" + val1 + "/Sportbook/InPlay"
url2 = "https://" + val2
browser.get(url)
browser2.get(url2)

#url - the url to fetch dynamic content from.
#delay - second for web view to wait
#block_name - id of the tag to be loaded as criteria for page loaded state.
def fetchHtmlForThePage(br):
    br.refresh()
    #grab the rendered HTML
    time.sleep(5)
    html = br.page_source
    #close the browser
    #browser.quit()
    #return html
    return html

while(1):
    #call the fetching function we created
    html = fetchHtmlForThePage(browser)
    html2 = fetchHtmlForThePage(browser2)
    #grab HTML document
    soup = BeautifulSoup(html,"lxml")
    soup2 = BeautifulSoup(html2,"lxml")
    #process it further as you wish.....
    ttime = soup.find_all('div',class_ = 'timer')
    strtime = ttime[int(row1)-1].string
    ttime2 = soup2.find_all('div',class_ = 'c-events__time')
    strtime2 = ttime2[int(row2)-1].span.string

    print(strtime)
    print(strtime2+'\n')

    if((int(strtime2[0:2]) - int(strtime[0:2])) >= 3 or 
       ((int(strtime2[0:2]) - int(strtime[0:2])) == 2 and (int(strtime2[3:5]) >= int(strtime[3:5])) )):
        frequency = 2500  # Set Frequency To 2500 Hertz
        duration = 500  # Set Duration To 1000 ms == 1 second
        winsound.Beep(frequency, duration)
        winsound.Beep(frequency, 750)
        winsound.Beep(frequency, 2*duration)
        winsound.Beep(frequency, 2*duration)
        now = datetime.now()
        print("Ekhtelafe zamani dar tarikhe: ", now)
        im = Image.open(r"image.png")  
        im.show()
        FROMADDR = "MYMAIL@gmail.com"
        LOGIN    = FROMADDR
        PASSWORD = "PASSWORD"
        TOADDRS  = ["MYMAIL@gmail.com"]
        SUBJECT  = "Done"

        msg = ("From: %s\r\nTo: %s\r\nSubject: %s\r\n\r\n"
               % (FROMADDR, ", ".join(TOADDRS), SUBJECT) )
        msg += "some text " + strtime + " " + strtime2 +"\r\n"

        try :
            server = smtplib.SMTP('smtp.gmail.com', 587)
            server.set_debuglevel(1)
            server.ehlo()
            server.starttls()
            server.login(LOGIN, PASSWORD)
            server.sendmail(FROMADDR, TOADDRS, msg)
            server.quit()
        except :
            print('nok')
            
        time.sleep(30)
        break
    
    time.sleep(10)

    
```
